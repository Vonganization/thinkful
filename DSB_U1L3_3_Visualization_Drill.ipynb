## Imports

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

## Import CSV file

```
# Import the csv file into a dataframe for analysis and manipulation

df = pd.read_csv('https://raw.githubusercontent.com/Vonganization/CSV_datasets/master/adult-training.csv', 
                 sep=",",
                 header=None,
                 names=['age','workclass','fnlwgt','education','education_num',
                        'marital_status',
                        'occupation',
                        'relationship',
                        'race',
                        'gender',
                        'capital_gain',
                        'capital_loss',
                        'hours_per_week',
                        'native_country',
                        'income_bracket'
                       ]
                )
df.head()
```

# Choose one variable and plot that variable four different ways.
 - Age

```
x = df['age']
f, axes = plt.subplots(2, 2, figsize=(10,10))
sns.distplot(x, color='red', ax=axes[0,0]);
sns.boxplot(x=None, y=x, color='blue', ax=axes[0,1]);
sns.kdeplot(x,color='green', shade=True, ax=axes[1,1]);
sns.pointplot(x=None, y=x, ax=axes[1,0]);
```

# Choose two continuous variables, and plot them three different ways.

## Import CSV file

```
df2 = pd.read_csv('https://raw.githubusercontent.com/Vonganization/CSV_datasets/master/weight-height.csv')
df2.head()
```

## Two continuous variables - Way One

```
sns.regplot(x='Height', y='Weight', data=df2, color='pink');
```

## Two continuous variables - Way Two

```
z1 = df2['Height']
z2 = df2['Weight']

sns.jointplot(z1 , z2, kind='kde', height=7, space=0)
```

## Two continuous variables - Way Three

```
sns.residplot(z1, z2, color='Green')
```

# Choose one continuous variable and one categorical variable, and plot them six different ways.

```
plt.figure(figsize=(20,5))
sns.violinplot(x='education', y='age', data=df, split=True, inner='quart');
```

```
plt.figure(figsize=(20,5))
sns.boxplot(x='education', y='age', data=df);
```

```
plt.figure(figsize=(20,5))
sns.barplot(x=df['education'],y=df['age']);
```

```
plt.figure(figsize=(20,5))
sns.pointplot(x=df['education'],y=df['age']);
```

```
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(x='age', y='education', data=df);
plt.subplot(1,2,2)
sns.boxplot(x='age', y='education', data=df);
```
